### 3.


Ile kobiet i ile mężczyzn wzięło udział w zmaganiach olimpijskich w poszczególnych latach?

```python
o = olimp.groupby('Year').Gender.value_counts()
o = o.reset_index()
m = o[o.Gender == 'Men']['count']
m = m.reset_index(drop=True)
m
```

```python
o = olimp.groupby('Year').Gender.value_counts()
o = o.reset_index()
o = o[o.Gender == 'Women']['count'] 
o = o.reset_index(drop=True)
o
```

```python
result = pd.concat([o, m], axis=1)
result
```

```python
x = olimp.groupby(["Gender","Year"]).size().unstack(0)
x
```

### 4.


Utwórz ramkę z procentowym udziałem kobiet względem poszczególnych lat, w których odbywały się igrzyska.

```python
x["% Women"] = (x.Women/(x.Men + x.Women) * 100)
x
```

### 5. 


Wyświetl ramkę 10 najbardziej utytułowanych medalistów zgodnie z rankingiem:
* brąz -- 1 punkt,
* srebro -- 2 punkty,
* złoto -- 6 punktów

```python
olimp.groupby(['Athlete', 'Medal']).size().unstack().fillna(0).astype('int')
```

```python
'Gold Silver Bronze Ranking'.split()
```

```python
df = olimp.groupby(['Athlete', 'Medal']).size().unstack().fillna(0).astype('int')
df['Ranking'] = 6*df.Gold + 2*df.Silver + df.Bronze
df.sort_values(by='Ranking', ascending=False, inplace=True)
df = df['Gold Silver Bronze Ranking'.split()]
df.head(10)
```
